# Load Finance PDF

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("IPO_SGTM.pdf")
documents = loader.load()

print(len(documents))  # number of pages
```

# Split text

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

docs = text_splitter.split_documents(documents)
print(len(docs))
```

# Create embeddings

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

# Store in FAISS (vector DB)

```python
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(docs, embeddings)
```

# Load the LLM

```python
from transformers import pipeline

# Load LLM
llm = pipeline(
    "text-generation",
    model="Qwen/Qwen2.5-1.5B",)
```

# RAG Query (Investing Q&A)

```python
def ask_finance_ai(question, k=3):
    docs = vectorstore.similarity_search(question, k=k)

    context = "\n".join([doc.page_content for doc in docs])

    prompt = f"""
You are a financial document analysis assistant.
Your task is to analyze IPO prospectus content.
Based ONLY on the provided document context, classify the stock as:
- BUY
- HOLD

You must justify the decision using financial indicators from the document.
If the information is insufficient, respond with HOLD.
Context:
{context}

Question:
{question}

Answer:
"""

    result = llm(prompt)
    return result[0]["generated_text"]
```

# Ask stock advisor questions

```python
print(ask_finance_ai("is it good to invest in this company"))
```
